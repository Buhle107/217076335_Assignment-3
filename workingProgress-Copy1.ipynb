```python
import pandas as pd

exams_df = pd.read_excel('C:\\Users\\buhle\\Downloads\\Data set\\exam scores.xlsx')
exams_df
```

```python
#Converting string to interger
exams_df['Gender'] = exams_df['Gender'].replace(['female'], '1')
exams_df['Gender'] = exams_df['Gender'].replace(['male'], '0')

exams_df['Race/Ethnicity'] = exams_df['Race/Ethnicity'].replace(['African', 'White', 'Coloured', 'Indian'], ['1', '2', '3','4'])
exams_df['Parental level of education'] = exams_df['Parental level of education'].replace(['High school', 'College', 'Diploma', 'Degree',"Master's Degree"], ['1', '2', '3','4','5'])

exams_df
```
